```python
import pandas as pd
cleaned_income_data=pd.read_excel('income_data.xlsx',index_col=0)
# cleaned_income_data.head()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(18,6), dpi=1600) 
alpha=alpha_scatterplot = 0.2 
alpha_bar_chart = 0.55
ax1 = plt.subplot2grid((2,3),(0,0))
# plots a bar graph of those who surived vs those who did not.               
cleaned_income_data.income_level.value_counts().plot(kind='bar', alpha=alpha_bar_chart)
# this nicely sets the margins in matplotlib to deal with a recent bug 1.3.1
ax1.set_xlim(-1, 2)
# puts a title on our graph
plt.title("Distribution of income level, (50000 = more income)") 

plt.subplot2grid((2,3),(0,1))
plt.scatter(cleaned_income_data.income_level, cleaned_income_data.age, alpha=alpha_scatterplot)
# sets the y axis lable
plt.ylabel("Age")
# formats the grid line style of our graphs                          
plt.grid(b=True, which='major', axis='y')  
plt.title("Income level by Age,  (50000 = More income)")
```

1)More people are in low income level and very few people are in high income level
2)From age graph age about 20 are having more income level
